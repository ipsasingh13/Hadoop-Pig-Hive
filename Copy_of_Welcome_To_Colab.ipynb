Q1. Write a Python program to perform basic arithmetic operations (addition, subtraction, multiplication, division, and modulus) on two numbers.
o	Prompt the user to enter two numbers.
o	Perform the arithmetic operations and print the results.
o	Use appropriate operators and print formatting for clear output.

```
# Prompt the user to enter two numbers
num1_str = input("Enter the first number: ")
num2_str = input("Enter the second number: ")

try:
    # Convert input strings to numbers (float for general arithmetic, int for modulus)
    num1 = float(num1_str)
    num2 = float(num2_str)

    # Perform arithmetic operations
    addition = num1 + num2
    subtraction = num1 - num2
    multiplication = num1 * num2

    print("\n--- Results --- ")
    print(f"{num1} + {num2} = {addition}")
    print(f"{num1} - {num2} = {subtraction}")
    print(f"{num1} * {num2} = {multiplication}")

    # Division: Handle division by zero
    if num2 != 0:
        division = num1 / num2
        print(f"{num1} / {num2} = {division}")
    else:
        print("Division by zero is not allowed.")

    # Modulus: Requires integer operands
    # Convert to integer for modulus, but be careful with non-integer inputs
    if num2 != 0:
        try:
            int_num1 = int(num1)
            int_num2 = int(num2)
            modulus = int_num1 % int_num2
            print(f"{int_num1} % {int_num2} = {modulus}")
        except ValueError:
            print("Modulus operation requires integer numbers. Could not perform modulus with given inputs.")
    else:
        print("Modulus by zero is not allowed.")

except ValueError:
    print("Invalid input. Please enter valid numbers.")
```

Q2.  Create variables of different data types (integer, float, string, boolean) and perform basic operations on them.
o	Assign values to variables of different data types.
o	Perform arithmetic operations on numeric data types.
o	Concatenate strings using the + operator.
o	Use logical operators to evaluate boolean expressions.

```
print("Q2. Create variables of different data types (integer, float, string, boolean) and perform basic operations on them.")

# Assign values to variables of different data types
my_integer = 10
my_float = 20.5
my_string = "Hello"
my_boolean = True

print(f"\nInteger variable: {my_integer}, Type: {type(my_integer)}")
print(f"Float variable: {my_float}, Type: {type(my_float)}")
print(f"String variable: {my_string}, Type: {type(my_string)}")
print(f"Boolean variable: {my_boolean}, Type: {type(my_boolean)}")

# Perform arithmetic operations on numeric data types
sum_result = my_integer + my_float
diff_result = my_float - my_integer
prod_result = my_integer * my_float
div_result = my_float / my_integer

print(f"\nArithmetic Operations:")
print(f"Sum (integer + float): {sum_result}")
print(f"Difference (float - integer): {diff_result}")
print(f"Product (integer * float): {prod_result}")
print(f"Division (float / integer): {div_result}")

# Concatenate strings using the + operator
second_string = " World!"
concatenated_string = my_string + second_string

print(f"\nString Concatenation:")
print(f"Concatenated string: {concatenated_string}")

# Use logical operators to evaluate boolean expressions
another_boolean = False

and_result = my_boolean and another_boolean
or_result = my_boolean or another_boolean
not_result = not my_boolean

print(f"\nLogical Operators:")
print(f"AND operation (True and False): {and_result}")
print(f"OR operation (True or False): {or_result}")
print(f"NOT operation (not True): {not_result}")
```

Q3. Write a program to take user input, process it, and display the result.
o	Prompt the user to enter their name.
o	Greet the user using their name.
o	Calculate and print the user's age based on their birth year.

```
from datetime import datetime

# Prompt the user to enter their name
user_name = input("Please enter your name: ")

# Greet the user using their name
print(f"Hello, {user_name}!")

# Prompt the user to enter their birth year
birth_year_str = input("Please enter your birth year (e.g., 1990): ")

try:
    birth_year = int(birth_year_str)
    current_year = datetime.now().year
    user_age = current_year - birth_year
    print(f"You are approximately {user_age} years old.")
except ValueError:
    print("Invalid input for birth year. Please enter a valid number.")
```

Q4.  Write a program to check if a number is even or odd.
o	Prompt the user to enter a number.
o	Use the modulus operator to determine if the number is even or odd.
o	Print the appropriate message.

```
# Prompt the user to enter a number
number_str = input("Please enter an integer: ")

try:
    number = int(number_str)
    # Use the modulus operator to determine if the number is even or odd
    if number % 2 == 0:
        print(f"The number {number} is even.")
    else:
        print(f"The number {number} is odd.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

Q5. Write a program to print the numbers from 1 to 10 using both for and while loops.
o	Use a for loop to iterate through a range of numbers.
o	Use a while loop with a counter variable.

```
# Using a for loop to print numbers from 1 to 10
print("Numbers from 1 to 10 using a for loop:")
for i in range(1, 11):
    print(i)

# Using a while loop to print numbers from 1 to 10
print("\nNumbers from 1 to 10 using a while loop:")
count = 1
while count <= 10:
    print(count)
    count += 1
```
